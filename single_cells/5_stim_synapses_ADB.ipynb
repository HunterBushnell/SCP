```python
################ stim_synapses functions ################
class AllenCell:
    
    ############### Cell Generation Functions ###############
    def __init__(self, gid, soma_diam_multiplier=1.0):
        # from allensdk.model.biophys_sim.config import Config
        # from allensdk.model.biophysical.utils import Utils
        # from neuron import h
        
        self._gid = gid
        self.synapses = []  # Keep track of all synapses
        self.netcons = []   # Keep track of NetCons
        self.stims = []     # Keep VecStims so they don't get garbage-collected
        self.vecs = []
        self.syn_locs = [] 
        
        description = Config().load('manifest.json')

        self.utils = Utils(description)
        self.h = self.utils.h
        self.Vinit = self.utils.description.data['conditions'][0]['v_init']
        # Cast all genome values to float
        for d in self.utils.description.data['genome']:
            if 'value' in d:
                d['value'] = float(d['value'])

        # Load morphology and parameters
        morphology_path = description.manifest.get_path('MORPHOLOGY')
        self.utils.generate_morphology(morphology_path.encode('ascii', 'ignore'))
        self.utils.load_cell_parameters()
        self.setup_morphology(soma_diam_multiplier)
        self._build_section_list()
    
    def setup_morphology(self,soma_diam_multiplier):

        self.soma = self.h.soma
        self.h.soma[0].diam *= soma_diam_multiplier
        self.dend = list(self.h.dend) if hasattr(self.h, 'dend') else []
        self.apic = list(self.h.apic) if hasattr(self.h, 'apic') else []
        self.axon = list(self.h.axon) if hasattr(self.h, 'axon') else []
        from neuron import h

        # Step 1: Set distance origin once (assuming soma(0.5) is your origin)
        h.distance(0, self.soma[0](0.5))

        # Step 2: Classify dendrites into proximal/distal
        self.proximal_dend_segs = []
        self.distal_dend_segs = []

        for sec in self.dend:
            for seg in sec:
                if 20 < h.distance(seg) < 100:
                    self.proximal_dend_segs.append(seg)  
                elif  h.distance(seg) >= 100:
                    self.distal_dend_segs.append(seg)

    def _build_section_list(self):
        # from neuron import h
        self.all = h.SectionList()
        for sec in h.allsec():
            self.all.append(sec)


############### Synapse Generation Functions ###############

    def _lognormal_mu_sigma(self, mean, std):
            """Return μ and σ for np.random.lognormal given arithmetic mean & std."""
            import math
            mu  = math.log(mean**2 / math.sqrt(std**2 + mean**2))
            sig = math.sqrt(math.log(1 + (std**2 / mean**2)))
            return mu, sig

######################### Generate Syns #########################


    def gen_n_synlocs(self, n_syns, dens_eq, seg_list):
        """
        Sample exactly n_syns synapse locations from seg_list using a desired
        density per µm vs. distance (dens_eq). Returns a list[Segment].

        Args
        ----
        n_syns   : int
        dens_eq  : number | callable(distance->density/µm) | array-like
        seg_list : list of NEURON Segments

        Returns
        -------
        all_syn_locs : list of Segments chosen with replacement
        """
        import numpy as np
        from neuron import h

        # RNG (reproducible if you set self.seed somewhere else)
        rng = getattr(self, "rng", None)
        if rng is None:
            self.rng = np.random.default_rng(getattr(self, "seed", None))
            rng = self.rng

        # Ensure distance origin (soma center)
        h.distance(0, self.soma[0](0.5))

        # Distances (µm) and per-segment lengths (µm)
        distances   = np.array([h.distance(seg) for seg in seg_list], dtype=float)
        seg_lengths = np.array([seg.sec.L / seg.sec.nseg for seg in seg_list], dtype=float)

        # ---- evaluate dens_eq safely to get density per µm at each distance ----
        n = len(seg_list)
        if dens_eq is None:
            dens = np.ones(n, dtype=float)
        elif callable(dens_eq):
            # try vector call; if shape/behavior is odd, fall back to scalar-map
            try:
                out = dens_eq(distances)
                if np.isscalar(out):
                    dens = np.full(n, float(out), dtype=float)
                else:
                    out = np.asarray(out, dtype=float)
                    if out.shape != distances.shape:
                        raise ValueError
                    dens = out
            except Exception:
                dens = np.array([float(dens_eq(float(d))) for d in distances], dtype=float)
        elif isinstance(dens_eq, (int, float, np.floating)):
            dens = np.full(n, float(dens_eq), dtype=float)
        else:
            dens = np.asarray(dens_eq, dtype=float)
            if dens.shape != distances.shape:
                raise ValueError(
                    "dens_eq provided as array-like must have same length as seg_list"
                )

        # No negative densities
        dens = np.clip(dens, 0.0, None)

        # ---- convert to sampling probabilities: p_i ∝ density(d_i) × segment_length_i ----
        raw = dens * seg_lengths
        s = raw.sum()
        if s <= 0:
            # fallback to length-only (uniform density)
            raw = np.clip(seg_lengths, 0.0, None)
            s = raw.sum()
            if s <= 0:
                raise ValueError("All segment lengths are zero; cannot sample.")

        p = raw / s

        # ---- sample with replacement to place exactly n_syns ----
        idx = rng.choice(len(seg_list), size=int(n_syns), replace=True, p=p)
        all_syn_locs = [seg_list[i] for i in idx]

        return all_syn_locs
    
    def gen_distr_synlocs(self, dens_eq, seg_list):

        all_syn_locs = []   # Or could use self.syn_locs.append(seg), instead of later when generating syns
        h.distance(0, self.soma[0](0.5))

        for seg in seg_list:
            seg_dist = h.distance(seg)                  # µm
            seg_len  = seg.sec.L / seg.sec.nseg         # µm

            syn_dens = dens_eq(seg_dist)                # Synapses per µm
            if syn_dens <= 0:                           # skip negative densities
                print('NEGATIVE/0 SYN_DENS, NO SYN CREATED!!!')
                continue

            n_syns = math.floor(syn_dens * seg_len)     #+ rng.random())
            if n_syns == 0:
                print(f'no syns created for seg {seg}: dist {seg_dist}, len {seg_len}, dens {syn_dens}')

            for syn in range(n_syns):
                    self.syn_locs.append(seg)
                    all_syn_locs.append(seg)

            # for _ in range(n_syn):
            #     # random position within the segment’s bounds
            #     #      segment runs from seg.x-Δx to seg.x+Δx
            #     dx = (rng.random() - 0.5) * (1/seg.sec.nseg)
            #     loc = seg.x + dx
            #     syn_locs.append(SynLoc(seg.sec, loc, seg_dist + dx*seg.sec.L))

        return all_syn_locs

    def gen_syn_locs(
            self,
            n_syns,
            dens_eq,        # =lamda d: -0.015*d + 4.25, = lamda: 2.0 
            seg_list,
            ):
        
        """ Generate synapse locations by density distribution dens_eq  """
        
        import math, re, random
        from neuron import h

        ### Generate Synaptic Locations ###
        ## If based on number and probability distrubution
        if n_syns is not None:
            all_syn_locs = self.gen_n_synlocs(n_syns, dens_eq, seg_list)

        ## If based on density distribution
        else:
            all_syn_locs = self.gen_distr_synlocs(dens_eq, seg_list)

        for seg in all_syn_locs:
            self.syn_locs.append(seg)
        return all_syn_locs

######################### Generate Syns Mechs #########################


    def draw_syn_wt(self, syn_params):
        """Generate a synaptic weight from a lognormal distribution."""

        wt_mean = syn_params.get('wt_mean', syn_params.get('initW', 0.001))
        wt_std  = syn_params.get('wt_std', 0.0) * wt_mean   #scaled off mean, std = 0  → fixed

        if wt_std > 0:
            syn_mu, syn_sig = self._lognormal_mu_sigma(wt_mean, wt_std)
            # mu  = math.log(wt_mean**2 / math.sqrt(wt_std**2 + wt_mean**2))
            # sig = math.sqrt(math.log(1 + (wt_std**2 / wt_mean**2)))
            draw_wt = lambda: np.random.lognormal(syn_mu, syn_sig)
        else:
            draw_wt = lambda: wt_mean

        syn_wt = draw_wt()

        return syn_wt

    def gen_syn_mechs(
            self,
            syn_loc,
            syn_params,
            ):
        """ Generate syn_name synapse with parameters syn_params. """

        # from neuron import h
        syn_type = syn_params['type']
        # print(f'syn_type: {type(syn_type)} | syn_loc: {syn_loc}')
        syn = getattr(h, syn_params['type'])(syn_loc)
        for param, val in syn_params.items():
            # if param in ('wt_mean', 'wt_std'): # Skip certain keys
            #     continue
            if hasattr(syn, param):
                setattr(syn, param, val)
        syn_wt = self.draw_syn_wt(syn_params)
        syn.initW = syn_wt
        # print(syn.initW)

        return syn

######################### Generate Spike Times #########################

    def homogeneous_poisson_timestamps(self, rate_hz, t_start, duration_ms):
        """
        Generate spike times for a homogeneous Poisson process.

        Parameters:
            rate_hz (float): desired firing rate in Hz.
            duration_ms (float): total time to generate over, in milliseconds.

        Returns:
            spike_times (list of float): spike timestamps in ms.
        """
        spike_times = []
        t = t_start         # Start of stimulation
        while t < duration_ms:
            isi = np.random.exponential(1000.0 / rate_hz)  # ISI in ms
            t += isi
            if t < duration_ms:
                spike_times.append(t)
        return spike_times


    def inhomogeneous_poisson_through_num_points(self,lambdas, win_length):
        t = np.zeros(len(lambdas) * win_length)
        lambdas = np.divide(lambdas,1000/win_length)
        
        for i, lambd in enumerate(lambdas):

            num_points = np.random.poisson(lambd)

            if num_points >= win_length:
                t[i * win_length : (i + 1) * win_length] = 1
                continue

            random_inds = np.random.choice(a = np.arange(win_length), size = num_points, replace = False)
            spikes = np.zeros(win_length)
            spikes[random_inds] = 1
            t[i * win_length : (i + 1) * win_length] = spikes

        return t
    

    def gen_spike_times(
            self,
            sim_params,
            syn_params,
            jitter_tstart,
    ):
        """ Generate spike trains (list of timestamps) for NetCon input. """

        # # Add lognormal randomness to start of input; could make sd auto from sim_params
        # t_mu, t_sig = self._lognormal_mu_sigma(sim_params['tstart'], 100)
        # tstart_dist = lambda: np.random.lognormal(t_mu, t_sig)
        # # sim_params['tstart'] = tstart_dist()
        # tstart_lognorm = tstart_dist()
        # print(tstart_lognorm)

        # If constant firing rate, specified in syn_params
        if isinstance(syn_params['freq'], int) or isinstance(syn_params['freq'], float):
            # t_mu, t_sig = self._lognormal_mu_sigma(200, 100)
            # tstart_dist = lambda: np.random.lognormal(t_mu, t_sig)
            # tstart_lognorm = tstart_dist()
            # # print(tstart_lognorm)
            if syn_params['delay'] is not None:
                jitter_tstart = syn_params['delay']
            spike_times = self.homogeneous_poisson_timestamps(
                            rate_hz=syn_params['freq'], 
                            # t_start=sim_params['tstart'],
                            t_start=jitter_tstart,
                            duration_ms=sim_params['tstop'])
        
        # If inhomogeneous based on bio data
        elif isinstance(syn_params['freq'], str):

            inhom_input_filepath = syn_params['freq']
            PFR = pd.read_csv(inhom_input_filepath,delimiter=",")
            bio_stim_input = np.array(PFR['AvgFiringRate'][PFR['Time'] >0])
            # bio_stim_input = PFR
            # print(type(bio_stim_input))

            spike_times = []

            # Generate baseline input until csv data stim
            baseline_factor = bio_stim_input[0]
            baseline_spike_times = self.homogeneous_poisson_timestamps(
                rate_hz=baseline_factor,
                # t_start=sim_params['tstart'],
                t_start=jitter_tstart,
                duration_ms=sim_params['delay']
            )
            for spk in baseline_spike_times:
                spike_times.append(spk)

            # Generate stim input from csv
            # bio_stim_input = [avg_fq - baseline_factor for avg_fq in bio_stim_input]
            stim_spikes = self.inhomogeneous_poisson_through_num_points(
                            bio_stim_input,
                            int(sim_params['bins']))
            time = np.arange(len(stim_spikes))
            # print(sim_params['delay']+len(time))
            stim_spike_times = time[stim_spikes==1]
            stim_spike_times = [spk + sim_params['delay'] for spk in stim_spike_times] # Add delay

            for spk in stim_spike_times:
                spike_times.append(spk)
            
            # Generate baseline input after csv data stim
            baseline_spike_times = self.homogeneous_poisson_timestamps(
                rate_hz=baseline_factor,
                t_start=sim_params['delay']+len(time),
                duration_ms=sim_params['tstop']
            )
            for spk in baseline_spike_times:
                spike_times.append(spk)

        elif isinstance(syn_params['freq'], dict):
            # print(syn_params)
            random_spktrn = random.randint(0,len(syn_params['freq']['base tune'])-1)
            input_spikes = syn_params['freq']['base tune'][random_spktrn]
            # print(f'spike train {random_spktrn} - spikes {spike_times}')

            spike_times = []
            for spk in input_spikes:
                # print(spk)
                spike_time = spk + 200 + 50 #diff in delays #sim_params['delay']
                # print(spike_time)
                spike_times.append(spike_time)


        else:
            print('Please use valid spike train input!')

        return spike_times

######################### Generate Syns #########################

    def gen_syns(
            self,
            syn_params,
            # bio_stim_input,
            sim_params,
            ):
        """ Generate syn_name synapses with parameters syn_params for segments in seg_list. """

        # from neuron import h

        syn_records = {}    # list of dicts we will optionally return
        syn_id = 0

        # # Add lognormal randomness to start of input; could make sd auto from sim_params
        t_mu, t_sig = self._lognormal_mu_sigma(sim_params['tstart'], 100)
        tstart_dist = lambda: np.random.lognormal(t_mu, t_sig)
        # sim_params['tstart'] = tstart_dist()
        jitter_start = tstart_dist()
        # print(jitter_start)
        # print(sim_params['tstart'])

        ### Generate synapses for each syn group in syn_params ###
        for syn_group in syn_params:
            if syn_params[syn_group]['N_syn'] is not None:
                if syn_params[syn_group]['N_syn'] < 1:
                    continue
                
            # Create record lsit for group
            syn_records[syn_group] = []

            # Get syn_params for group
            syn_params_group = syn_params[syn_group]

            # Determine which dendrite sections for group syns
            syn_segs = syn_params_group['segs']
            if syn_segs == 'all':
                seg_list = self.proximal_dend_segs + self.distal_dend_segs
            elif syn_segs == 'proximal':
                seg_list = self.proximal_dend_segs
            elif syn_segs == 'distal':
                seg_list = self.distal_dend_segs
            elif syn_segs == 'soma':
                seg_list = self.soma
            else:
                print('NO DENDRITE SECTIONS SELECTED!!!')


            ### Generate synaptic locations ###
            all_syn_locs = self.gen_syn_locs(
                                n_syns = syn_params_group['N_syn'], 
                                dens_eq = syn_params_group['dist_func'], 
                                seg_list = seg_list,
                                )


            for syn_loc in all_syn_locs:
                ### Generate synaptic mechanisms ###
                syn = self.gen_syn_mechs(syn_loc, syn_params_group)
                self.synapses.append(syn)

                ### Generate synaptic input (spike trains) ###
                spike_times = self.gen_spike_times(
                                    sim_params, 
                                    syn_params_group,
                                    jitter_start,
                                    )

                ### Generate NetCon & Stimulation ###
                vec = h.Vector(spike_times)
                stim = h.VecStim()
                stim.play(vec)
                nc = h.NetCon(stim, syn)
                nc.weight[0] = 1 
                
                # self.synapses.append(syn)
                self.vecs.append(vec)
                self.stims.append(stim)
                self.netcons.append(nc)


                ### Generate synaptic record ###
                syn_records[syn_group].append({    # Could add more, or auto from syn_params_group/h.
                    "syn_id": syn_id,
                    "group": syn_group,
                    "type": syn_params_group['type'],
                    "weight": syn.initW,
                    "distance": h.distance(syn_loc),
                    "section": syn_loc.sec.name(),
                    "x": syn_loc.x,
                    "spike_times": spike_times,
                    # "NMDA_ratio": nmda_wt, # EXAMPLE, FOR FUTURE
                })
                # print(f'synapse {syn_id} generated: {syn_records[syn_group][-1]}')
                syn_id = syn_id + 1
                
            print(f'{syn_group} synapses generated: {len(all_syn_locs)}')

        return syn_records



    def __str__(self):
        return f"AllenCell(soma={self.soma}, dendrites={len(self.dend)})"



############ Other cell building functions #############
```

```python
import numpy as np
import pandas as pd
import json
from pathlib import Path


# ---------- low-level helpers ----------

def _generate_homogeneous_poisson_trains(
    rate_hz: float,
    t_start_ms: float,
    t_end_ms: float,
    n_syn: int,
    rng: np.random.Generator,
):
    """
    Homogeneous Poisson trains in [t_start_ms, t_end_ms].
    """
    window_ms = float(t_end_ms) - float(t_start_ms)
    if n_syn <= 0 or window_ms <= 0.0 or rate_hz <= 0.0:
        return [np.array([], dtype=float) for _ in range(max(int(n_syn), 0))]

    window_s = window_ms / 1000.0
    lam = float(rate_hz)

    trains = []
    for _ in range(int(n_syn)):
        t = 0.0
        times_ms = []
        while True:
            isi = rng.exponential(1.0 / lam)  # seconds
            t += isi
            if t > window_s:
                break
            times_ms.append(t_start_ms + t * 1000.0)
        trains.append(np.array(times_ms, dtype=float))

    return trains


def _generate_inhomogeneous_trains_from_curve(
    times_ms,
    rates_hz,
    n_syn,
    rng,
    t_start_ms,
    t_end_ms,
):
    """
    Inhomogeneous Poisson via thinning from a rate(t) curve.

    times_ms, rates_hz: 1D arrays (ms, Hz), monotonic times.
    Generated spikes are in [t_start_ms, t_end_ms].
    """
    times_ms = np.asarray(times_ms, dtype=float)
    rates_hz = np.asarray(rates_hz, dtype=float)

    if times_ms.ndim != 1 or rates_hz.ndim != 1 or times_ms.size != rates_hz.size:
        raise ValueError("times_ms and rates_hz must be 1D arrays of same length")

    # Restrict curve to the requested window (in sim coordinates)
    mask = (times_ms >= t_start_ms) & (times_ms <= t_end_ms)
    if not np.any(mask):
        return [np.array([], dtype=float) for _ in range(max(int(n_syn), 0))]

    curve_t = times_ms[mask]
    curve_r = rates_hz[mask]
    r_max = float(curve_r.max())
    if r_max <= 0.0 or n_syn <= 0:
        return [np.array([], dtype=float) for _ in range(max(int(n_syn), 0))]

    window_s = (t_end_ms - t_start_ms) / 1000.0

    trains = []
    for _ in range(int(n_syn)):
        # Homogeneous Poisson at r_max over [0, window_s]
        t = 0.0
        cand_times = []
        while True:
            isi = rng.exponential(1.0 / r_max)
            t += isi
            if t > window_s:
                break
            cand_times.append(t_start_ms + t * 1000.0)

        cand_times = np.asarray(cand_times, dtype=float)
        if cand_times.size == 0:
            trains.append(cand_times)
            continue

        # Thin with r(t)/r_max
        rt = np.interp(cand_times, curve_t, curve_r)
        accept_prob = rt / r_max
        u = rng.random(size=accept_prob.shape)
        keep = u < accept_prob
        trains.append(cand_times[keep])

    return trains


def trains_to_serializable(trains):
    return [np.asarray(tr, dtype=float).tolist() for tr in trains]


def save_trains_to_json(trains, path):
    path = Path(path)
    data = {"trains": trains_to_serializable(trains)}
    with path.open("w", encoding="utf-8") as f:
        json.dump(data, f, indent=2)
    return path


# ---------- high-level helper: bio curve + baseline ----------

def build_bio_trains_with_baseline_from_csv(
    csv_path,
    sim_params,
    n_syn,
    rng=None,
):
    """
    Replicates the functional structure of your old gen_spike_times()
    for the bio-stim case:

      - load CSV with columns 'Time' and 'AvgFiringRate'
      - use AvgFiringRate[Time>0] as bio_stim_input
      - baseline rate = first value of that array
      - baseline before: [tstart, delay]
      - inhomogeneous stim: aligned to delay, using the bio curve
      - baseline after: [delay + stim_duration, tstop]

    Returns
    -------
    trains : list of np.ndarray
        One spike-train per synapse (ms, sim time).
    """
    if rng is None:
        rng = np.random.default_rng()

    csv_path = Path(csv_path)
    PFR = pd.read_csv(csv_path, delimiter=",")

    # Match your old code: use Time > 0 segment
    mask = PFR["Time"] > 0
    time_local = PFR.loc[mask, "Time"].to_numpy(dtype=float)  # starts near 0
    bio_stim_input = PFR.loc[mask, "AvgFiringRate"].to_numpy(dtype=float)

    if bio_stim_input.size == 0:
        raise ValueError("Bio-stim CSV has no Time>0 samples.")

    baseline_factor = float(bio_stim_input[0])

    tstart = float(sim_params["tstart"])
    delay = float(sim_params["delay"])
    tstop = float(sim_params["tstop"])

    # Duration of the bio-stim segment in ms (local domain)
    stim_duration = float(time_local[-1] - time_local[0])
    stim_start = delay
    stim_end = delay + stim_duration

    # Pre-baseline: [tstart, delay]
    pre_trains = _generate_homogeneous_poisson_trains(
        rate_hz=baseline_factor,
        t_start_ms=tstart,
        t_end_ms=delay,
        n_syn=n_syn,
        rng=rng,
    )

    # Inhomogeneous stim segment: curve shifted to start at 'delay'
    times_ms = time_local - time_local[0] + delay
    stim_trains = _generate_inhomogeneous_trains_from_curve(
        times_ms=times_ms,
        rates_hz=bio_stim_input,
        n_syn=n_syn,
        rng=rng,
        t_start_ms=stim_start,
        t_end_ms=stim_end,
    )

    # Post-baseline: [stim_end, tstop]
    post_trains = _generate_homogeneous_poisson_trains(
        rate_hz=baseline_factor,
        t_start_ms=stim_end,
        t_end_ms=tstop,
        n_syn=n_syn,
        rng=rng,
    )

    # Combine per-synapse
    trains = []
    for i in range(int(n_syn)):
        pre = pre_trains[i]
        stim = stim_trains[i]
        post = post_trains[i]
        if pre.size or stim.size or post.size:
            merged = np.sort(np.concatenate([pre, stim, post]))
        else:
            merged = np.array([], dtype=float)
        trains.append(merged)

    return trains
```
